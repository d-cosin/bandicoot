# Importing Relevant Libraries

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

# Defining the Equation

```python
def rho_equation(rho=0, initial_state=0, error_term_range=[-0.5, 0.5], size_error_vector=250):
    x = []
    
    # assign the initial state
    x.append(initial_state)
    
    error_terms = np.random.uniform(error_term_range[0], error_term_range[1], size_error_vector)
    
    for idx, error_term in enumerate(error_terms):
        next_x = rho*x[idx] + error_term
        x.append(next_x)
    
    return x
```

```python
x = rho_equation()
plt.figure(figsize=(10,5))
plt.plot(x)
plt.ylim(-1, 1);
```

```python
x = rho_equation(rho=0.5)
plt.figure(figsize=(10,5))
plt.plot(x)
plt.ylim(-1, 1);
```

```python
x = rho_equation(rho=1.0)
plt.figure(figsize=(10,5))
plt.plot(x);
```
